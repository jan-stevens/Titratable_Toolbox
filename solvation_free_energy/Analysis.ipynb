```python
import matplotlib.pyplot as plt
import numpy as np

# Load in Data

## Define Bead types
Bead_Type                    = np.array([ "C1",  "C4", "C4h", "N4a",  "X2", "SC2", "SC3", "SC5", "SC6", "SN4a", "SX2", "TC2", "TC4", "TC5", "TN3d"])
Bead_Adapted                 = [ 1, 2, 4, 5, 6, 7, 8, 9, 10, 11, 14]

## Solvation Free Energy for pre release Martini v3.0.0 and the corresponding titratable version
FE_Martini_Pre                   = -np.array([8.77, 5.34, 5.41, -5.44, 5.34, 5.93, 5.23, 3.72, 2.02, -3.58, 5.21, 7.48, 6.10, 4.36, -2.11])
FE_Titratable_Martini_Pre_acidic = -np.array([8.23, 4.72, 4.84, -5.43, 4.90, 5.22, 4.34, 2.74, 1.06, -4.11, 4.57, 6.64, 5.30, 3.50, -2.70])
FE_Titratable_Martini_Pre_basic  = -np.array([8.23, 4.53, 4.62, -6.47, 4.45, 5.75, 4.80, 3.23, 1.45, -4.30, 5.05, 7.34, 6.00, 4.14, -2.63])

## Solvation Free Energy for Martini v3.0.0 and the corresponding titratable version
FE_Martini_v300              = np.array([-8.73, -4.78, -4.91,  5.49, -3.58, -6.16, -5.38, -3.72, -2.01,   3.75, -3.74, -7.33, -6.11, -4.29,   2.62])
FE_Titratable_Martini_acidic = np.array([-8.39, -4.49, -4.59,  5.39, -3.30, -5.28, -4.56, -3.01, -1.27,   4.40, -2.97, -6.39, -5.28, -3.56,   3.25])
FE_Titratable_Martini_basic  = np.array([-8.21, -4.25, -4.30,  6.53, -3.08, -5.82, -5.04, -3.41, -1.54,   4.47, -3.40, -7.17, -5.97, -4.12,   3.04])

# Create masks
S_Beads = np.array([type.startswith("S") for type in Bead_Type]) 
T_Beads = np.array([type.startswith("T") for type in Bead_Type]) 
R_Beads =  ~(T_Beads | S_Beads)
```

```python

# Create plot
fig, ax = plt.subplots(1,2, figsize=(24,8))

ax[0].plot(Bead_Type, FE_Martini_v300, linewidth=2, alpha=0.9, label='Martini v3.0.0')
ax[0].plot(Bead_Type, FE_Titratable_Martini_basic, linewidth=2, alpha=0.9, label='Titratable Martini: pH7.5')
ax[0].plot(Bead_Type, FE_Titratable_Martini_acidic, linewidth=2, alpha=0.9, label='Titratable Martini: pH3.0')
ax[0].set_xlabel("Bead Types", fontsize=25, labelpad=65)
ax[0].set_ylabel("Free Energy [$kJ\cdot mol^{-1}$]", fontsize=25, labelpad=13)
ax[0].set_title('Solvation free energies of neutral Martini v3.0.0 beads', fontsize = 25, pad=15)

# Add annotation arrows
ax[0].annotate('R-Beads', xy=(0.177, -0.08), xytext=(0.177, -0.15), xycoords='axes fraction', fontsize=20, color='black',
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=5.4, lengthB=.5', lw=2.0, color='black'))
ax[0].annotate('S-Beads', xy=(0.532, -0.08), xytext=(0.532, -0.15), xycoords='axes fraction', fontsize=20, color='black',
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=7.1, lengthB=.5', lw=2.0, color='black'))
ax[0].annotate('T-Beads', xy=(0.862, -0.08), xytext=(0.862, -0.15), xycoords='axes fraction', fontsize=20, color='black', 
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=4.7, lengthB=.5', lw=2.0, color='black'))

# Nice layout
ax[0].grid(True, color='gainsboro', linestyle='-', linewidth=0.5)
ax[0].set_facecolor('white')
ax[0].set_ylim(-9, 8.9)
ax[0].tick_params(axis='x', labelsize=15)
ax[0].tick_params(axis='y', labelsize=15)
for tick in Bead_Adapted:
    ax[0].get_xticklabels()[tick].set_color('red')
ax[0].legend(fontsize='xx-large')


ax[1].plot(Bead_Type, FE_Martini_Pre, linewidth=2, alpha=0.9, label='Martini3 v3.0.0')
ax[1].plot(Bead_Type, FE_Titratable_Martini_Pre_basic, linewidth=2, alpha=0.9, label='Titratable Martini : pH7.5')
ax[1].plot(Bead_Type, FE_Titratable_Martini_Pre_acidic, linewidth=2, alpha=0.9, label='Titratable Martini : pH3.0')
ax[1].set_xlabel("Bead Types", fontsize=25, labelpad=65)
ax[1].set_ylabel("Free Energy [$kJ\cdot mol^{-1}$]", fontsize=25, labelpad=13)
ax[1].set_title('Solvation free energies of neutral Martini v3.0.0 beads', fontsize = 25, pad=15)

# Add annotation arrows
ax[1].annotate('R-Beads', xy=(0.177, -0.08), xytext=(0.177, -0.15), xycoords='axes fraction', fontsize=20, color='black',
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=5.4, lengthB=.5', lw=2.0, color='black'))
ax[1].annotate('S-Beads', xy=(0.532, -0.08), xytext=(0.532, -0.15), xycoords='axes fraction', fontsize=20, color='black',
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=7.1, lengthB=.5', lw=2.0, color='black'))
ax[1].annotate('T-Beads', xy=(0.862, -0.08), xytext=(0.862, -0.15), xycoords='axes fraction', fontsize=20, color='black', 
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=4.7, lengthB=.5', lw=2.0, color='black'))

# Nice layout
ax[1].grid(True, color='gainsboro', linestyle='-', linewidth=0.5)
ax[1].set_facecolor('white')
ax[1].set_ylim(-9, 8.9)
ax[1].tick_params(axis='x', labelsize=15)
ax[1].tick_params(axis='y', labelsize=15)
for tick in Bead_Adapted:
    ax[1].get_xticklabels()[tick].set_color('red')
ax[1].legend(fontsize='xx-large')

plt.tight_layout()
plt.show()
```

```python
FE_Var_Acidic = FE_Titratable_Martini_acidic - FE_Martini_v300
FE_Var_Basic = FE_Titratable_Martini_basic - FE_Martini_v300

# Create plot
fig, ax = plt.subplots(1,1, figsize=(12,8))

ax.plot(Bead_Type, FE_Var_Acidic, linewidth=2, alpha=0.9, label='Titratable Martini: pH3.0')
ax.plot(Bead_Type, FE_Var_Basic, linewidth=2, alpha=0.9, label='Titratable Martini: pH7.5')
ax.set_xlabel("Bead Types", fontsize=25, labelpad=65)
ax.set_ylabel("$\Delta$ Free Energy [$kJ\cdot mol^{-1}$]", fontsize=25, labelpad=13)
ax.set_title('Deviations of Solvation Free Energy', fontsize = 25, pad=15)


# Add annotation arrows
ax.annotate('R-Beads', xy=(0.177, -0.08), xytext=(0.177, -0.15), xycoords='axes fraction', fontsize=20, color='black',
         ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=5.4, lengthB=.5', lw=2.0, color='black'))
ax.annotate('S-Beads', xy=(0.532, -0.08), xytext=(0.532, -0.15), xycoords='axes fraction', fontsize=20, color='black',
         ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=7.1, lengthB=.5', lw=2.0, color='black'))
ax.annotate('T-Beads', xy=(0.862, -0.08), xytext=(0.862, -0.15), xycoords='axes fraction', fontsize=20, color='black', 
            ha='center', va='bottom', arrowprops=dict(arrowstyle='-[, widthB=4.7, lengthB=.5', lw=2.0, color='black'))


# Nice layout
ax.grid(True, color='gainsboro', linestyle='-', linewidth=0.5)
ax.set_facecolor('white')
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.tight_layout()
plt.legend(fontsize='xx-large')
plt.show()
```

```python
# Define needed variables
Beads = [R_Beads, S_Beads, T_Beads]
Titles = ["R Beads", "S Beads", "T Beads"]
FE_Var_Acidic = FE_Titratable_Martini_acidic - FE_Martini_v300
FE_Var_Basic = FE_Titratable_Martini_basic - FE_Martini_v300

# Create plot
fig, axs = plt.subplots(1, len(Beads), figsize=(18,7), sharey=True)

# Iterate over subplots
for i in range(len(Beads)):

    avg = (np.average(FE_Var_Acidic[Beads[i]]) + np.average(FE_Var_Basic[Beads[i]])) / 2
    axs[i].plot(Bead_Type[Beads[i]], FE_Var_Acidic[Beads[i]], linewidth=2, alpha=0.9, label='Titratable Martini: pH3.0')
    axs[i].plot(Bead_Type[Beads[i]], FE_Var_Basic[Beads[i]], linewidth=2, alpha=0.9, label='Titratable Martini: pH7.5')
    axs[i].hlines(y=avg, xmin=0, xmax=len(Bead_Type[Beads[i]]) - 1, colors='k', linestyles='--', lw=2)
    axs[i].set_title(Titles[i], fontsize = 20, pad=15)
    axs[i].grid(True, color='gainsboro', linestyle='-', linewidth=0.5)
    axs[i].tick_params(axis='x', labelsize=15)
    axs[i].tick_params(axis='y', labelsize=15)
    axs[i].set_facecolor('white')

axs[0].legend(fontsize='large')
axs[0].set_ylabel("$\Delta$ Free Energy [$kJ\cdot mol^{-1}$]", fontsize=25, labelpad=15)
fig.suptitle('Deviations of Solvation Free Energy', fontsize = 25, y=1.05)
fig.supxlabel('Bead Types', fontsize = 25, y=-0.05)
plt.show()
```
